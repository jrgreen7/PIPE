```python
from Bio import SeqIO
import re

fasta_sequences = SeqIO.parse(open('./data/uniprot-proteome UP000002311.fasta'),'fasta')
with open('./data/uniprot-proteome UP000002311_stripped.fasta', 'w') as out_file:
    for fasta in fasta_sequences:
        name, sequence = fasta.id, str(fasta.seq)
        new_name = re.findall(r'\|(.*?)\|', name)[0]
        # print(new_name)
        fasta.id = new_name
        fasta.name = new_name
        fasta.description = new_name
        print(sequence)
        # print(fasta)
        SeqIO.write(fasta, out_file, "fasta")
fasta_sequences
```
